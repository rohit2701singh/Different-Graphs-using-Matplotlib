## np.random.seed(fixed num):
If you do not set the seed, the NumPy random number generator will produce different sequences of random numbers every time you run your program

```python
# without np.random.seed()
print(np.random.randint(0,10,5))
np.random.randint(0,10,5)
```

```python
# with np.random.seed()
np.random.seed(42)
np.random.randint(0,10,5)
```

```python
np.random.seed(42)
np.random.randint(0,10,5)
```

```python
# Fixing random state for reproducibility

# A string representation of a float value in [0, 1] inclusive for gray level (e.g., '0.5')
# refers to a way of specifying a shade of gray using a single value that ranges from 0 (black) to 1 (white).
# 's' parameter specifies the size of each data point, the 'c' parameter specifies the color of each data point,
# and the alpha parameter controls the transparency of the data points.

np.random.seed(42)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
# print(colors)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

# **Annotation**
Annotations in Matplotlib are text or arrowed lines that point to specific data points or locations on a plot.

```python
# simple example

x = [1, 2, 3, 4]
y = [5, 6, 7, 8]

plt.scatter(x, y, marker='o', s=250)
plt.text(1,5, 'point 1', fontdict={'size':12, 'color': 'red'})
plt.text(2,6, 'point 2')
plt.text(3,7, 'point 3')
plt.text(4,8, 'point 4')
```

```python
batter = pd.read_csv('/content/drive/MyDrive/dataset for matplotlib learning from youtube/Copy of batter.csv')
print(batter.shape)
batter.sample(5)
```

```python
# s= jintna jyda runs utna bada size
# random_state parameter is used to set the seed.
# This ensures that you get the same random sample every time you run the code with the same random_state value.

sample_batter_df = batter.head(100).sample(25, random_state=5)

plt.figure(figsize=(15, 6))
plt.scatter(sample_batter_df['avg'], sample_batter_df['strike_rate'], s=sample_batter_df['runs'], alpha=0.6)
plt.xlabel('avg')
plt.ylabel('strike rate')
plt.tight_layout()

for i in range(sample_batter_df.shape[0]):
    plt.text(sample_batter_df['avg'].values[i], sample_batter_df['strike_rate'].values[i], sample_batter_df['batter'].values[i])
```

## A different way to plot graph

plt.subplots() is used to create a new figure (fig) and a single axis (ax) within that figure. The ax object is what you use to plot your data or customize the plot further.

```python
plt.subplots()
# output: (<Figure size 640x480 with 1 Axes>, <Axes: >)
```

# **3D Contour Plot**

Contour plots also called level plots are a tool for doing multivariate analysis and visualizing 3-D plots in 2-D space.


```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(x,y)
z = xx**2 + yy**2

fig = plt.figure(figsize=(12, 6))
ax = plt.subplot()

graph = ax.contour(xx,yy,z, cmap='viridis')
fig.colorbar(graph)
fig.show()
```

## filled contour: contourf

```python
# This image is similar to the first example of a surface graph. Imagine you are flying over a mountain;
# you will be able to see where the peak is and where it is deepest.
# Yellow color showing peak and violet color showing depth

x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(x,y)
z = xx**2 + yy**2

fig = plt.figure(figsize=(12, 6))
ax = plt.subplot()

graph = ax.contourf(xx,yy,z, cmap='viridis')
fig.colorbar(graph)
fig.show()
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(x,y)
z= np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12, 6))
ax = plt.subplot()

graph = ax.contourf(xx,yy,z, cmap='viridis')
fig.colorbar(graph)
fig.show()
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(x, y)
z = xx**2 + yy**2

fig = plt.figure()

# Create a 3D subplot in the first position of a 2x2 grid
ax1 = fig.add_subplot(2, 2, 1, projection='3d')
graph = ax1.plot_surface(xx, yy, z, cmap='viridis')

ax2 = fig.add_subplot(2, 2, 2)
graph = ax2.contour(xx, yy, z, cmap='viridis')
fig.colorbar(graph, pad=0.1)

ax3 = fig.add_subplot(2, 2, 3)
graph = ax3.contourf(xx, yy, z, cmap='viridis')
fig.colorbar(graph, pad=0.1)

fig.tight_layout()
```
